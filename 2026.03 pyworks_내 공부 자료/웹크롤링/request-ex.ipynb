```python
import requests
from bs4 import BeautifulSoup

url = "https://www.seoul.go.kr/main/index.jsp"

response = requests.get(url)

# beautifulsoup 객체 생성
soup = BeautifulSoup(response.text, "html.parser")

print(soup.title)   
print(soup.title.text)   # 태그 안의 텍스트만 추출
```

```python
# 메뉴 제목 추출 - select_one() : CSS 선택자 이용, select() : CSS 선택자 이용하여 여러 요소 추출
first_li = soup.select_one("div.m_service > ul > li") 

print(first_li.text)  
```

```python
# 모든 메뉴 제목 추출 - select() : CSS 선택자 이용하여 여러 요소 추출
all_li = soup.select("div.m_service > ul > li")
print(all_li)  

for li in all_li:
    print(li.text)
```
