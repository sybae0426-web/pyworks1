### 판다스의 자료구조 - 시리즈 관리
#### 데이터가 순차적으로 나열된 1차원 배열의 형태를 가진다.
#### '#'의 개수가 많을수록 글꼴이 작아진다 - 마크다운(Markdown) 문서: 확장자(.md)

```python
import pandas as pd  #별칭(alias) - pd로 사용

data = ['키보드', '마우스', 'USB 메모리']
index = [1, 2, 3]     # 0번부터 시작하지 않고 1부터 시작하는거 세팅

# s1 객체 생성
s1 = pd.Series(data, index)
print(s1)

# s1의 속성
print(s1.index)        
print(s1.values)      # ['키보드' '마우스' 'USB 메모리']
print(s1.shape)       # 튜플(3, )  데이터의 형태 -> 1차원 3개의 요소

# 요소 선택
print(s1[1])
print(s1[2])
print(s1[3])
```

```python
# 두번째 시리즈
data = [20000, 32000, 10000]
index = [1, 2, 3]


s2 = pd.Series(data, index)
print(s2)
```

### 데이터프레임 : 행과 열로 이루어진 2차원 표 형태
#### 데이터프레임의 data는 파이썬의 딕셔너리 자료구조로 만든다.

```python
import pandas as pd

# 데이터프레임 자료 생성(객체: df)
df = pd.DataFrame({
    "name": s1,        # 첫번째 시리즈
    "price" : s2       # 두번째 시리즈
})

print(df)

# 데이터프레임의 속성
print(df.columns)         # 칼럼     - Index(['name', 'price'], dtype='object')
print(df.shape)           # 자료형태 - (3, 2)


# 칼럼 검색
print("-------------------------")
print("name")
print(df["name"])
print("-------------------------")
print("price")
print(df["price"])
```
