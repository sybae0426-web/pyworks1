```python
import pandas as pd

# datas/food.cvs 파일 읽기
df = pd.read_csv("datas/food.csv")
print(df)
# 데이터 탐색
df.info() 

# 결측치 확인
df.insa().sum()
```

```python
# 판다스의 통계, 수학 관련 함수
print("개수:", df["메뉴"].count())
print("평균:", df["가격"].mean())             # 가격의 평균
print("평균:", round(df["가격"].mean(), 2))   # 반올림
print("분산:", df["가격"].var())              # 분산
# 편차 - 가격에서 평균을 뺀값
print("표준편차:", round(df["가격"].std()))       # 표준편차
print("최대값:", df["가격"].max())             # 가격의 최대
print("최소값:", df["가격"].min())             # 가격의 최소
print("최대값 인덱스:", df["가격"].idxmax())          # 가격의 최대값의 위치(인덱스)
print("최소값 인덱스:", df["가격"].idxmin())          # 가격의 최소값의 위치(인덱스)

# 사분위 수 - quantile()
print("1사분위수:", df["가격"].quantile(.25)) # (0.25 뜻임)
print("2사분위수:", df["가격"].quantile(.50)) # (50% 뜻임)
print("3사분위수:", df["가격"].quantile(.75)) # (75% 뜻임)
print("4사분위수:", df["가격"].quantile(1.0)) # (100% 뜻임)

print("---------------------------------------------------------------------")
# 조건 - 1사분위 수보다 작은 가격의 음식 정보를 출력
result = df["가격"] < df["가격"].quantile(.25)
print(df[result])

print("---------------------------------------------------------------------")
# 최빈값 - mode()
freq_menu = df["메뉴"].mode()[0]
print(f"메뉴의 최빈값: {freq_menu}")

freq_class = df["분류"].mode()[0]
print(f"분류의 최빈값: {freq_class}")
```

```python
import pandas as pd
food = pd.read_csv("datas/food.csv")
print(food)
print("------------------------------")

# 그룹화 - groupby()
# 메뉴별 가격의 평균
food.groupby("메뉴")['가격'].mean()

# 분류별 가격의 평균
food.groupby("분류")['가격'].mean()
food.groupby("분류")['할인가'].mean()

# 수량 칼럼 추가
food['수량'] = [20, 10, 15, 5, 7, 7, 8]
food

# food2.csv 파일을 생성
food.to_csv('datas/food2.csv', index=False)

# food2.csv 파일 읽기
df = pd.read_csv('datas/food2.csv')
df


# 메뉴별 수량의 합계를 출력
df.groupby("메뉴")['수량'].sum()
```
